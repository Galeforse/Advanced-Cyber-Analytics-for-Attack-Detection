```r
setwd("D://Datasets")

set.seed(7)
x <- c(rep(c(0,3), times=8), 11, 13, rep(c(0,3), times=7))
x <- x+5
```

```r
# png(file="BasicAnomaly2.png")
plot(x, col=ifelse(x<=10, 'DarkBlue', 'Red3'), type='b', cex=ifelse(x<=10, 1.6, 2.4), pch=19, 
  ylim = c(5, 18), main="Example of anomalies detected in practice", xlab="Event", ylab="Height of intensity")
lines(x, col=ifelse(x>=10, 'Blue', 'Black') )
# dev.off()
```
